```python
# import SAGA code

import SAGA
from SAGA import ObjectCuts as C
print('SAGA code version', SAGA.__version__)
```

```python
from easyquery import Query
```

```python
saga_database = SAGA.Database(root_dir='/home/yymao/Dropbox/Academia/Collaborations/SAGA')
saga_obj_catalog = SAGA.ObjectCatalog(saga_database)
```

```python
saga_galaxies = saga_obj_catalog.load(has_spec=True, cuts=(C.is_clean & C.is_galaxy))
```

```python
lowz_galaxies = Query('SPEC_Z < 0.05').filter(saga_galaxies)
```
